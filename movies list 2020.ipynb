```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('movie_list2020.csv')
```

```python
df.head()
```

```python
sns.set()
sns.displot(data=df, x='imdb rating', kind='kde')
```

```python
sns.displot(data=df, x='length', kind='kde')
```

```python
top_rated_films = df.loc[df['imdb rating'] > 7.6]
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=top_rated_films, x='Title', y='imdb rating')
plt.xticks(rotation = -30)
```

```python
worst_rated_films = df.loc[df['imdb rating'] < 2.5]
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=worst_rated_films, x='Title', y='imdb rating')
plt.xticks(rotation = -30)
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Month')
```
